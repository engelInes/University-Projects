```python
#task 6
#unfair probability spinner
import numpy as np
import matplotlib.pyplot as plt
def spinner(probabilityDistribution):
    randomNumber=np.random.uniform(0,1)
    index=0
    while randomNumber>0:
        if randomNumber - probabilityDistribution[index]>0:
            randomNumber=randomNumber-probabilityDistribution[index]
            index=index+1
        else:
            return index

def simulatingSpinnerTrials(probabilityDistribution, noTrials):
    outcomesList=[]
    for _ in range(noTrials):
        currentOutcome=spinner(probabilityDistribution)
        outcomesList.append(currentOutcome)
    return outcomesList

def plotOutcomes(outcomesList):
    
    unique_outcomes, counts = np.unique(outcomesList, return_counts=True)
    labels = unique_outcomes
    
    plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title('Pie Chart of Spinner Outcomes')
    plt.show()


probabilityDistribution1 = [0.2, 0.5, 0.1, 0.2]
noTrials = 10000
outcomesList = simulatingSpinnerTrials(probabilityDistribution1, noTrials)
plotOutcomes(outcomesList)
```

```python
#unfair probability spinner but instead of uniform dist the input is a list of nat numbers
#each number holds how many degrees has the center angle 
import numpy as np
import matplotlib.pyplot as plt
def spinner(angles):
    total_angles = sum(angles)
    randomNumber = np.random.uniform(0, 1) * total_angles
    cumulative_sum = 0
    
    for index, angle in enumerate(angles):
        cumulative_sum += angle
        if randomNumber <= cumulative_sum:
            return index

def simulatingSpinnerTrials(angles, noTrials):
    outcomesList = []
    for _ in range(noTrials):
        currentOutcome = spinner(angles)
        outcomesList.append(currentOutcome)
    return outcomesList

def plotOutcomes(outcomesList):
    unique_outcomes, counts = np.unique(outcomesList, return_counts=True)
    labels = unique_outcomes
    
    plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title('Pie Chart of Spinner Outcomes')
    plt.show()

angles = [30, 60, 15, 30] 
noTrials = 10000
outcomesList = simulatingSpinnerTrials(angles, noTrials)
plotOutcomes(outcomesList)
```

```python
#Task 7
#pipe of radius 30cm rectangular wall of l=100cm
#monte carlo simulation to compute the prob the ball falls inside the pipe
import numpy as np
import matplotlib.pyplot as plt
pipeRadius=0.3
rectangularLength=1
ballRadius=0.1
trialsList=10000

def getBallPosition():
    x=np.random.uniform(-1/2,1/2)
    y=np.random.uniform(-1/2,1/2)
    return x,y

def checkIfInPipe(x,y):
    if(x**2+y**2<=pipeRadius**2):
        return 1
    return 0

def throwBalls(trialsList):
    cnt=0
    for i in range(0,trialsList):
        x,y=getBallPosition()
        cnt=cnt+checkIfInPipe(x,y)
    return cnt

cntInside=throwBalls(trialsList)
result=cntInside/trialsList
print(result)

mathematicalProbability=(np.pi * pipeRadius * pipeRadius) / rectangularLength**2
print(mathematicalProbability)
```
