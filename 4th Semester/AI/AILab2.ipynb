```python
#exercise 2
from sklearn.model_selection import train_test_split
df = pd.read_csv('score.csv')

X = np.array(df['Hours'])
y = np.array(df['Scores'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
X = X.reshape(-1, 1)#from 1D to 2D

X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression #relationship between a dependent variable and one or more independent variables
lr = LinearRegression().fit(X_train, y_train)

print(f"Linear Regression-Training set score: {lr.score(X_train, y_train):.2f}")
print(f"Linear Regression-Test set score: {lr.score(X_test, y_test):.2f}")
```

```python
lr.predict([[7], [15], [31]])
```

```python
plt.scatter(X_train, y_train, s=300)
plt.plot(X_train, lr.predict(X_train))

plt.title('Training set')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
plt.scatter(X_test, y_test, s=300)
plt.plot(X_test, lr.predict(X_test))

plt.title('Test set')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
plt.scatter(X, y, s=300)
plt.plot(X, lr.predict(X))

plt.title('Predictions')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
#perform all three regularizations
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.7).fit(X_train, y_train)

print("L2 regularization (Ridge)")
print(f"Ridge Regression-Training set score: {ridge.score(X_train, y_train):.2f}")
print(f"Ridge Regression-Test set score: {ridge.score(X_test, y_test):.2f}\n")
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=1.0).fit(X_train, y_train)

print("L1 regularization (Lasso)")
print(f"Lasso Regression-Training set score: {lasso.score(X_train, y_train):.2f}")
print(f"Lasso Regression-Test set score: {lasso.score(X_test, y_test):.2f}\n")
```

```python
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet(alpha=0.01, l1_ratio=0.01).fit(X_train, y_train)

print("Elastic Net")
print(f"Elastic Net-Training set score: {elastic_net.score(X_train, y_train):.2f}")
print(f"Elastic Net-Test set score: {elastic_net.score(X_test, y_test):.2f}")
```
